```python
#Import needed class and function
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
#Read the dataset
dataset=pd.read_csv('Real estate.csv')
dataset
```

```python
dataset.columns
```

```python
dataset["Y house price of unit area"].value_counts()
```

```python
#Input and output variables
indepVariable=dataset[['X1 transaction date', 'X2 house age',
       'X3 distance to the nearest MRT station',
       'X4 number of convenience stores', 'X5 latitude', 'X6 longitude']]
depVariable=dataset[['Y house price of unit area']]
```

```python
indepVariable
```

```python
depVariable
```

```python
#Split test and train set
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(indepVariable,depVariable,test_size=1/3,random_state=0)
```

```python
#Model Creation
from sklearn.svm import SVR
regressor=SVR()
regressor=regressor.fit(X_train,y_train)
```

```python
#Evaluation Metrics
y_pred=regressor.predict(X_test)
y_pred
```

```python
from sklearn.metrics import r2_score
r_score=r2_score(y_test,y_pred)
r_score
```
